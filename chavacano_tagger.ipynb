```python
filepath = '/sample_spanish.txt'
text = open(filepath, encoding='utf-8').read()
document = nlp(text)
#for token in document:
    #print(token, token.pos_)

token_dict = Counter()
#Calculating number of tokens in Spanish sample
for token in document:
  if str(token) not in token_dict.keys():
    token_dict[str(token)] = 1
  else:
    token_dict[str(token)] += 1
  #if str(token) in particles:
    #token.pos_ = "PART"

print(token_dict.keys())

tokens = 0
types = len(token_dict.keys())
for token in token_dict.keys():
  tokens += token_dict[token]

print(f'Number of tokens: {tokens}')
print(f'Number of types: {types}')

#Calculating most common types
most_common_types = token_dict.most_common(20)

print(f'Most common types')
for type, freq in most_common_types:
  print(f'{type} : {freq}')

tags = [token.pos_ for token in document]
tag_counts = {}
for tag in tags:
    if tag not in tag_counts:
        tag_counts[tag] = 0
    tag_counts[tag] += 1
most_common_tag = max(tag_counts, key=tag_counts.get)
print(most_common_tag)
print(tag_counts)
tag_options = list(tag_counts.keys())
tag_options.remove('SPACE')
print(tag_options)
```

```python
results = '/spanish_sample_tagged_results.txt'
predictions = '/spanish_sample_tagged_gold.txt'
from collections import Counter

def calc_accuracy(results, predictions):
  results_file = open(results, 'r')
  preds_file = open(predictions, 'r')
  result_list = results_file.readlines()
  preds_list = preds_file.readlines()
  final_result = []
  final_pred = []
  prop_noun_dict = {}
  for line in result_list:
    line = line[:-1]
    tags = line.split()
    final_result.append(tags)
  for line in preds_list:
    line = line[:-1]
    tags = line.split()
    final_pred.append(tags)
  #print(final_result)
  total = 0
  correct = 0
  propn_mistake = []
  i = 0
  while i < len(final_result):
    if len(final_result[i]) == len(final_pred[i]):
      j = 0
      while j < len(final_result[i]):
        if final_result[i][j] == final_pred[i][j]:
          correct += 1
        elif final_result[i][j] == 'PROPN':
          propn_mistake.append(final_pred[i][j])
        total += 1
        j += 1
    else:
      total += len(final_pred[i])
      #print(i, final_result[i])
      #print(i, final_pred[i])
    i += 1
  print('Total number of tags:', total)
  print('Total correct', correct)
  print(f'Accuracy: {correct/total}%')
  counter = Counter(propn_mistake)
  print(counter)

calc_accuracy(results, predictions)
```
